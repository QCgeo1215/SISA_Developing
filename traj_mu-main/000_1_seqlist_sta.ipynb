```python
# 函数：提取列表的第一个值作为基准值
def get_first_value(lst_str):
    try:
        lst = ast.literal_eval(lst_str) if isinstance(lst_str, str) else lst_str
        return lst[0] if len(lst) > 0 else None
    except:
        return None

# 获取基准值
df['lat_grid_sta'] = df['lat_grid'].apply(get_first_value)
df['lng_grid_sta'] = df['lng_grid'].apply(get_first_value)

# 找出所有需要处理的列
lat_cols = [col for col in df.columns if col.startswith('lat_grid_') and col != 'lat_grid_sta']
lng_cols = [col for col in df.columns if col.startswith('lng_grid_') and col != 'lng_grid_sta']

# 对每个需要处理的列进行计算
for col in lat_cols:
    new_col = f"{col}-sta"
    df[new_col] = df['lat_grid_sta'] - df[col].apply(lambda x: x[0] if isinstance(x, list) else x)

for col in lng_cols:
    new_col = f"{col}-sta"
    df[new_col] = df['lng_grid_sta'] - df[col].apply(lambda x: x[0] if isinstance(x, list) else x)

# 显示结果
print(df.head(10))
```

```python
# 读取CSV文件
file_path = '3-Slidng_Windows_Data_Clustering_Clean-sta.csv'
df.to_csv(file_path, index=False)
#df= pd.read_csv(file_path)
```
